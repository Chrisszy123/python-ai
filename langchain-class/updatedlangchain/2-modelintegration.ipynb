### model integration with OPENAI, GOOGLE GEMINI, AND GROQ

```python
import os

from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("gpt-4.1")
```

```python
response = model.invoke("What is the weather in London?")
response
```

### STREAMING AND BATCH


Streaming is a way to send the llm response as it it being generated.. instead of waiting for the generation to be complete, this improves user experience, especially for large content

```python
model.stream("What is the weather in London?")
```

```python
for chunk in model.stream("What is the weather in London?"):
    print(chunk)
```

### BATCH

Batch is used to handle parallel calls

```python
model.batch(
    [ "What is the weather in London?",
    "What is the weather in New York?"],
    config={
        "max_concurrency": 5 # Optional: Limit the number of concurrent requests
    }
)
```
